```r
years <- subset(aggregate(price ~ year, data = myDF, FUN = mean))$year
```

```r
plot(years, prices)
```

```r
plot(years[prices<1500000], prices[prices<1500000])
```

```r
plot(years[prices<100000], prices[prices<100000])
```
