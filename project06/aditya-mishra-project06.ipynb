```r
#5A.
myDF <- read.csv("/anvil/projects/tdm/data/olympics/athlete_events.csv", stringsAsFactors=TRUE)
```

```r
plot(tapply(myDF$Height, myDF$Weight, mean), xlab = "Weight", ylab = "Mean height according to weight")
```

```r
#5B.
dotchart(tapply(myDF$Age[myDF$Sex == "M"], myDF$Race[myDF$Sex == "M"], mean), xlab = "Age in years", ylab = "Race code")
```

```r
dotchart(tapply(myDF$Age[myDF$Sex == "F"], myDF$Race[myDF$Sex == "F"], mean), xlab = "Age in years", ylab = "Race code")
```

5A. We can see that the heights increase sharply with weight initially, and reach a plateau with high variance for the higher weights. The heights have a much lower variance, nearly a continuous curve for the lower weight classes.

5B. We can see that the females of every race live longer than the males of every race by an almost constant difference of about 5 years. It is probably due to the males' tendencies of being more risk prone, like taking up a hazardous occupation.
